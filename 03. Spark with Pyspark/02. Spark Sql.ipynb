```
clientes = spark.read.parquet("dbfs:/FileStore/tables/arquivos/Atividades/Clientes.parquet")

itensVendas = spark.read.parquet("dbfs:/FileStore/tables/arquivos/Atividades/ItensVendas.parquet")

produtos = spark.read.parquet("dbfs:/FileStore/tables/arquivos/Atividades/Produtos.parquet")
                                
vendas = spark.read.parquet("dbfs:/FileStore/tables/arquivos/Atividades/Vendas.parquet")    

vendedores= spark.read.parquet("dbfs:/FileStore/tables/arquivos/Atividades/Vendedores.parquet")   
```

```
spark.sql('create database VendasVarejo')
```

```
spark.sql('use VendasVarejo')
```

```
clientes.write.saveAsTable('clientes')

vendas.write.saveAsTable('vendas')

itensVendas.write.saveAsTable('itensVendas')

produtos.write.saveAsTable('produtos')

vendedores.write.saveAsTable('vendedores')
```

```
spark.sql('show tables').display()
```

```
vendas_consulta = vendas.join(itensVendas, vendas.VendasID == itensVendas.VendasID,'inner').drop(itensVendas.VendasID)
```

```
produtos.show()
```

```
vendas_consulta = vendas_consulta.join(produtos, vendas_consulta.ProdutoID == produtos.ProdutoID,'inner').drop(produtos.ProdutoID)
```

```
vendas_consulta.show()
```

```
clientes.show()
```

```
vendas_consulta = vendas_consulta.join(clientes, vendas_consulta.ClienteID == clientes.ClienteID,'inner').drop(clientes.ClienteID)
```

```
vendas_consulta.select(func.col('Cliente'),func.col('Data'),func.col('Produto'),func.col('VendedorID'),func.col('ValorTotal')).show()
```

```
vendas_consulta.createOrReplaceTempView('vendas_consulta')
```
