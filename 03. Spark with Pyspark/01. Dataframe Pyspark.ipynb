```
clientesDF = spark.read.parquet("dbfs:/FileStore/tables/arquivos/Atividades/Clientes.parquet")

itensVendasDF = spark.read.parquet("dbfs:/FileStore/tables/arquivos/Atividades/ItensVendas.parquet")

produtosDF = spark.read.parquet("dbfs:/FileStore/tables/arquivos/Atividades/Produtos.parquet")
                                
vendasDF = spark.read.parquet("dbfs:/FileStore/tables/arquivos/Atividades/Vendas.parquet")    

vendedoresDF= spark.read.parquet("dbfs:/FileStore/tables/arquivos/Atividades/Vendedores.parquet")   
```

```
clientesDF.select(func.col('Cliente'),func.col('Estado'),func.col('Status')).show()
```

```
clientesDF.select(func.col('Cliente'),func.col('Estado'),func.col('Status')).filter((func.col('Status') == 'Platinum') | (func.col('Status') == 'Gold')).orderBy(func.col('Status'),func.col('Cliente')).show()
```

```
cliente_vendas = clientesDF.join(vendasDF,vendasDF.ClienteID ==  clientesDF.ClienteID).orderBy(func.col('Status'))
```

```
cliente_vendas.show()
```

```
analise = cliente_vendas.select('Status').groupBy('Status').count().withColumnRenamed('count','TotalVendas%')
```

```
analise.show()
```

```
analise = analise.select(func.col('Status'),func.col('TotalVendas%')/540*100,).withColumnRenamed('((TotalVendas% / 540) * 100)','TotalVendas%')
```

```
analise = analise.select(func.col('Status'),func.round(analise["TotalVendas%"], 2)).withColumnRenamed('round(TotalVendas%, 2)','TotalVendas%')
```

```
analise = analise.orderBy(func.col('TotalVendas%').desc())
```
